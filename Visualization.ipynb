###  Cheapest Restaurant in Each City

```
%sql
WITH city_cheapest AS (
  SELECT city, name, cost,
         ROW_NUMBER() OVER (PARTITION BY city ORDER BY cost ASC) AS rank
  FROM swiggy_gold
)
SELECT city, name, cost
FROM city_cheapest
WHERE rank = 1
```

### Top Restaurants by Max Combos per City

```
%sql
WITH city_max_combos AS (
  SELECT city, name, MAX(size(menu_items)) AS max_combos
  FROM swiggy_gold
  GROUP BY city, name
)
SELECT city, name, max_combos
FROM city_max_combos
ORDER BY max_combos DESC
LIMIT 20
```

### No of restaurants in each rating category

```
%sql
--total restaurants in each ratings
SELECT CASE 
         WHEN Rating >= 4.1 THEN '4.1-5.0'
         WHEN Rating >= 3.1 THEN '3.1-4.0'
         WHEN Rating >= 2.1 THEN '2.1-3.0'
         ELSE 'Below 2.1'
       END AS Rating_Category,
       COUNT(*) AS Count
FROM swiggy_gold
where rating >0
GROUP BY Rating_Category
```

### Most Popular Restaurant Chains - Franchises

### Restaurants that are highly rated yet affordable

```
%sql
--rank restaurants that are highly rated yet affordable.
SELECT name, ROUND(Rating, 2) AS Rating ,Cost, City,
DENSE_RANK() OVER (ORDER BY round(Rating, 2) DESC, cost asc) AS Dense_Rank
FROM swiggy_gold
WHERE Rating > 4.5 AND Cost < 200;
```

### Avg cost for each city


```
%sql
--Avg cost for each city
SELECT City, AVG(Cost) AS Average_Cost
FROM swiggy_gold
GROUP BY City
ORDER BY Average_Cost DESC;
```

### Most popular cuisine in each city

```
%sql
--most popular cuisine in each city
WITH CuisinePopularity AS (
    SELECT 
        city,
        cuisine,
        COUNT(*) AS cuisine_count
    FROM 
        swiggy_gold
    GROUP BY 
        city, cuisine
),
RankedCuisines AS (
    SELECT 
        city,
        cuisine,
        cuisine_count,
        RANK() OVER (PARTITION BY city ORDER BY cuisine_count DESC) AS rank
    FROM 
        CuisinePopularity
)
SELECT 
    city,
    cuisine AS most_popular_cuisine
FROM 
    RankedCuisines
WHERE 
    rank = 1
ORDER BY 
    city;
```

### Average price for veg and non-veg

```
%sql
SELECT mi.Veg_or_Non_Veg, round(AVG(mi.Price),2) AS Avg_Price
FROM swiggy_gold
LATERAL VIEW explode(menu_items) AS mi
GROUP BY mi.Veg_or_Non_Veg;
```

### Most Popular Item

```
%sql
--most popular item
SELECT mi.Item_Name, COUNT(*) AS Item_Count
FROM swiggy_gold
LATERAL VIEW explode(menu_items) AS mi
GROUP BY mi.Item_Name
ORDER BY Item_Count DESC
limit 100
```

### Number of restaurant in each price category in each city

```
%sql
--number of restaurant in each price category in each city
SELECT City,
       CASE 
         WHEN Cost<= 150 THEN 'Under 150'
         WHEN Cost <=300 THEN '151 to 300'
         WHEN Cost <=500 THEN '301 to 500'
         ELSE 'Above 500'
       END AS Price_Category,
       COUNT(*) AS Restaurant_Count
FROM swiggy_gold
GROUP BY City, Price_Category
ORDER BY City,Restaurant_Count;
```
